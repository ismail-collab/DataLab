<h2><u> EXERCICE 2 </u></h2>

```python
x=np.linspace(0,10,100)
f=3*x+5*np.cos(3*x)
plt.plot(x,f,label="f(x)")
plt.xlabel("Les x")
plt.ylabel("f(x)")
plt.legend()
plt.title("La fonction f(x) ")
#affichage de graphique 
plt.show()
```

```python
x=np.linspace(0,10,100)
g=np.tan(x+2)
plt.plot(x,g,label="g(x)")
plt.xlabel("Les x")
plt.ylabel("g(x)")
plt.legend()
plt.title("La fonction g(x) ")
#affichage de graphique 
plt.show()
```

```python
x=np.linspace(0,10,100)
h=np.exp(3*x+3)
plt.plot(x,h,"r--",label="h(x) ")
plt.xlabel("Les x")
plt.ylabel("h(x)")
plt.legend()
plt.title("La fonction h(x) ")
#affichage de graphique 
plt.show()
```
