<h1 style ="color:green">afficher quelque colonnes(liste [country,continent,total_deaths]) de la dataframe dans un df1 sous forme </h1>

```python
#afficher quelque colonnes de la dataframe 
#construire une liste 
#preciser le nom des variables que je vais travailler avec
L=['country','continent','total_deaths']
#ki yejbedhom n7othom fi dataframe jdida df1 
df1=df_covid[L]#ejbedly mel data frame 3 hekom
#printi df1
df1
#df_covid_vaccination[L]
#print(df_covid_vaccination[['country','continent','total_deaths']])
```

<h1 style ="color:green">afficher les pays dans le taux de décés est <1000 </h1>

```python
#afficher les pays dans le taux de décés est <1000
print(df1[df1['total_deaths']<1000])
```

<h1 style ="color:green">nombre total des pays dont le taux de deces est <1000 </h1>

```python
#nombre total des pays dont le taux de deces est <1000
df2=df1[df1['total_deaths']<1000]
print(len(df2))
#print(len(df[df['total_deaths']<1000]))
df2.shape
```

<h1 style ="color:green">reindexation dans une nv df4</h1>

```python
df4=df1.groupby('continent').mean().reset_index()
#reset indices to match new data frame
#to reformat the data frame indicies 
#bech ywali mafamech decalage 
df4
```

<h1 style ="color:green">rep graphique continent total_deaths df4</h1>

```python
from matplotlib import pyplot as plt
```

<h1 style ="color:green">création de graphique avec la méthode plot</h1>

```python
#création de graphique avec la méthode plot 
plt.plot(df4['continent'],df4['total_deaths'],c="red")
plt.xlabel("continent")
plt.ylabel("total deaths")
plt.title("means deaths per continent")
plt.legend(["mean deaths"],loc="upper left")
#affichage de graphique 
plt.show()
```
